```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn
import sklearn.datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
```

```python
# Load The Boston Housing Dataset
dataset = sklearn.datasets.load_boston()
```

```python
print(dataset)
```

```python
# Load the dataset to a Pandas DataFrame
dataframe = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
# Print the first 5 rows of the DataFrame
dataframe.head()
```

```python
# Add the PRICE column to the DataFrame
dataframe['PRICE'] = dataset.target
```

```python
# Print the first 5 rows of the DataFrame
dataframe.head()
```

```python
# Shape of the DataFrame
dataframe.shape
```

```python
# Check for any missing values
dataframe.isnull().sum()
```

```python
# Summary statistics of the DataFrame
dataframe.describe()
```

```python
# Construct a heatmap to visualize the correlation
correlation = dataframe.corr()
plt.figure(figsize=(10,10))
seaborn.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
# Drop the PRICE column from the DataFrame
X = dataframe.drop(['PRICE'], axis=1)

# Store PRICE column in the variable Y
Y = dataframe['PRICE']
```

```python
print(X)
print(Y)
```

```python
# Split the data into a train and test set
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
# Load the model
model = XGBRegressor()
```

```python
# Train the model
model.fit(x_train, y_train)
```

```python
# Accuracy of training data prediction
trainPrediction = model.predict(x_train)
```

```python
print(trainPrediction)
```

```python
# R Squared Error
rSquaredError = metrics.r2_score(y_train, trainPrediction)

# Mean Absolute Error
meanAbsoluteError = metrics.mean_absolute_error(y_train, trainPrediction)
```

```python
print("R Squared Error:", rSquaredError)
print("Mean Absolute Error:", meanAbsoluteError)
```

```python
# Accuracy of test data prediction
testPrediction = model.predict(x_test)
```

```python
# R Squared Error
rSquaredError = metrics.r2_score(y_test, testPrediction)

# Mean Absolute Error
meanAbsoluteError = metrics.mean_absolute_error(y_test, testPrediction)
```

```python
print("R Squared Error:", rSquaredError)
print("Mean Absolute Error:", meanAbsoluteError)
```
